```
from scipy import stats

print(stats.mode([1 , 1 , 2 , 2])[0][0])
```

```
import pandas as pd
```

```
Df1 = pd.read_csv("/content/submission_digit_9972.csv")
```

```
Df1.head()
```

```
Labels1 = list(Df1.label)
```

```
Df2 = pd.read_csv("/content/submission_digit_9969.csv")
Labels2 = list(Df2.label)
```

```
Df3 = pd.read_csv("/content/submission_99714.csv")
Labels3 = list(Df3.label)
```

```
pred_Label = []

for i , j , k in zip(Labels1 , Labels2 , Labels3):
    pred_Label.append([i , j , k])
```

```
pred_Label[0]
```

```
new_predictions = []

for i in pred_Label:
    new_predictions.append(stats.mode(i)[0][0])
```

```
import pandas as pd

sub_dic = {'filename' : list(Df1.filename) , 'label' : new_predictions}

Data = pd.DataFrame.from_dict(sub_dic)
```

```
Data.to_csv("submission_digit.csv" , header = True , index = False)
```
