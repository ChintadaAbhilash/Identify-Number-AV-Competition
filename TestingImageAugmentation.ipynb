```
import skimage
import cv2 

def Rotate(img):
    angle = randint(-20 , 20)
    img = cv2.imread(img)
    img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
    img = skimage.transform.rotate(img, angle, resize=False, center=None, order=1, mode='constant')
    img = cv2.resize(img , (64 , 64))
    return img
```

```
img = Rotate("/content/pexels-photo-414612.jpeg")
```

```
import matplotlib.pyplot as plt 

plt.imshow(img)
plt.show()
```

**Scale**

```
def Scale(img):
    img = cv2.imread(img)
    img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
    img = skimage.transform.rescale(img, 0.25, anti_aliasing=False)
    img = cv2.resize(img , (64 , 64))
    return img
```

```
img = Scale("/content/pexels-photo-414612.jpeg")

import matplotlib.pyplot as plt 

plt.imshow(img)
plt.show()
```
